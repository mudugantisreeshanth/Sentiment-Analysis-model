```python
import torch
if torch.cuda.is_available():
    device = 0
    print("running on gpu")
else:
  device=-1
  print("running on  cpu")
```

```python
#pipeline -> high level tool of hugging face that handles complicated stuff

from transformers import pipeline
print("Loading sentiment model..")
classifier=pipeline("sentiment-analysis",device=device)
print("Model loaded successfully")
```

```python
#test the model with positive review
result=classifier("This is a bad masterclass")
print(result)
```

```python
#batching  ->listing of sentence
results=classifier(["I cant believe  how good easy this is",
                    "I am finding this topic very easy, I love it",
                    "This video is 1hr long"])
print(results)
```

```python
#text generation
#GEN AI
#google/flan-t5-large

print("Loading text generation model..")
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="gpt2"
)
```

```python
prompt="Explain rag in simple terms"

result=pipe(prompt,max_new_tokens=100)
print("\n--ai response---")
print(result)
```

```python
#emotion aware ai
#plan
#Taking user from user
#Implementing sentiment analysis by classifier
#based on sentiment ,we will give a secret prompt using if else to our generator
#generate custom reply
#show it to the user
print("Emotion Aware AI assisstant...")
print("Hello ! I am AI assisstant , How are u feeling today?")
user_input=input(">")
sentiment=classifier(user_input)[0]
print(sentiment)
label=sentiment["label"]
score=sentiment["score"]
print(f"(Debug: {label} sentiment with {score*100:.2f}% confidence)")
if label=="NEGATIVE":
  prompt=f"""
  The user is feeling bad.Their exact words were:"{user_input}"
  Write a short,kind,empathetic reply. Acknoledge their feeling and offer support.
  """
elif label=="POSITIVE":
  prompt=f"""
  The user is feeling good.Their exact words were:"{user_input}"
  Write a short,enthusiastic,happy reply that celebrates their good mood.
  """
else:
  prompt:f"""
  The user made neutral statement.their exact words were:"{user_input}"
  write a short,polite and neutral reply acknowledging their statement.
  """
print("Generating reply...")

generated_reply=pipe(prompt,max_new_tokens=80)
print(generated_reply[0]["generated_text"])
```
